```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import chart_studio.plotly as pl
import plotly.offline as of
import cufflinks as cf
import datetime as dt
import folium as fo
import plotly.graph_objs as go
%matplotlib inline
```

```
of.init_notebook_mode(connected = True)
cf.go_offline()
```

```
def configure_plotly_browser_state():
  import IPython
  display(IPython.core.display.HTML('''
        <script src="/static/components/requirejs/require.js"></script>
        <script>
          requirejs.config({
            paths: {
              base: '/static/base',
              plotly: 'https://cdn.plot.ly/plotly-latest.min.js?noext',
            },
          });
        </script>
        '''))
```

```
url="https://raw.githubusercontent.com/nytimes/covid-19-data/master/us-counties.csv"
data = pd.read_csv(url)
```

```
url1="https://raw.githubusercontent.com/yashgarg-7/hello-world/master/data.csv"
data5=pd.read_csv(url1)
```

```
data5=data5.drop(['State'], axis=1)
data5.head()
```

```
data=data.drop(['fips'], axis=1)
data.head()
```

```
states4= data.groupby(['state','county'])['cases'].max().sort_values(ascending=False)
states4.head()
```

```
states5=states4.to_frame()
states5.head()
```

```
states6= states5.groupby(['state'])['cases'].sum().sort_values(ascending= False)
states6.head()
```

```
configure_plotly_browser_state()
states6.iplot(kind='bar', xTitle='States', yTitle='No. of cases ', title='Statewise Distribution of COVID-19 in the USA')
```

```
states7= states5.groupby(['county'])['cases'].sum()
states7
```

```
configure_plotly_browser_state()
states7.iplot(kind='scatter', xTitle='counties of different states', yTitle='No. of cases', title="Countywise distribution of COVID-19 in US states",gridcolor='grey', mode='markers',size=5 )
```

```
data1=data.drop(['state','county','deaths'], axis=1)
data1
```

```
data1['date']=pd.to_datetime(data1['date'])
series=data.groupby(data1['date'].dt.date).sum()
series.head()
```

```
configure_plotly_browser_state()
series.iplot(kind='box')
```

```
configure_plotly_browser_state()
series.iplot(kind='surface')
```

```
configure_plotly_browser_state()
series.iplot(kind='bar', xTitle="Dates", yTitle="No. of cases/deaths")
```

```
data2=states6.to_frame()
data2.reset_index(inplace=True)
data2.head()
```

```
data6=data2.sort_values('state')
data6=data6.reset_index()
data6.head()
```

```
data7=data6.drop(['index'],axis=1)
data8=data7.drop([11,36,41,49])
data8=data8.reset_index()
data8.head()
```

```
data9=data8.drop(['index'], axis=1)
data9.head()
```

```
data3=pd.concat([data9, data5], axis=1)
data3.head()
```

```
data4=dict(type='choropleth',
           locations=data3['Abbreviation'], 
           locationmode='USA-states',
           z= data3['cases'],
           text=data3['state'])
```

```
layout=dict(geo={'scope':'usa'})
```

```
p=go.Figure(data=data4, layout=layout)
```

```
configure_plotly_browser_state()
of.iplot(p)
```
